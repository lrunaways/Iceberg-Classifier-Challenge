```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from random import randint
import tensorflow as tf
from sklearn.model_selection import StratifiedKFold
import lightgbm as lgb
%matplotlib inline

#Train data
data_train = pd.read_json('E://kaggle//statoil//train.json')
data_train['inc_angle'] = pd.to_numeric(data_train['inc_angle'],errors='coerce')

X_band1=np.array([np.array(band).astype(np.float32) for band in data_train["band_1"]])
X_band2=np.array([np.array(band).astype(np.float32) for band in data_train["band_2"]])
X_band3 = (X_band1+X_band2)/2#
X_all = np.concatenate([X_band1[:, :, np.newaxis], X_band2[:, :, np.newaxis], X_band3[:,:,np.newaxis]], axis=-1)#
X_all = np.reshape(X_all, (1604,75,75,3))
print("X_all:", X_all.shape)

inc_angle_all = np.array(data_train["inc_angle"]).reshape(data_train.shape[0],1)
y_all = np.array(data_train["is_iceberg"]).reshape(data_train.shape[0],1)
from sklearn.model_selection import train_test_split
num_test = 0.2 #randomly shuffle this data into four variables. In this case,
                #I'm training 80% of the data, then testing against the other 20%.
X_train_all, X_test_all, y_train_all, y_test_all, inc_angle_train_all, inc_angle_test_all = train_test_split(X_all,
                                                            y_all, inc_angle_all,test_size=num_test, random_state = 5)
print(X_train_all.shape, y_train_all.shape, X_test_all.shape, y_test_all.shape,
      inc_angle_train_all.shape, inc_angle_test_all.shape)
```

# Keras

```python
#Import Keras.
from matplotlib import pyplot
from keras import backend as K
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Input, Flatten, Activation
from keras.layers import GlobalMaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.merge import Concatenate
from keras import initializers
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, Callback, EarlyStopping, ReduceLROnPlateau, TensorBoard
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K
import gc
from PIL import Image

def print_fold(n):
    print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")
    print("$$$$$$$$$$$$$$$$$$$", n, " FOLD $$$$$$$$$$$$$$$$$$$$$")
    print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")

def plot_history(history_):
    # summarize history for loss
    plt.plot(history_.history['loss'][1:])
    plt.plot(history_.history['val_loss'][1:])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    
def print_stats(stat_name, array):
    print("\n")
    print(stat_name, ": \n\t", array)
    print("mean", stat_name, ": ", array.mean())
    print("std", stat_name, ": ", array.std())

def get_callbacks(filepath, patience=2):
    es = EarlyStopping('val_loss', patience=patience, mode="min")
    msave = ModelCheckpoint(filepath, save_best_only=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, epsilon=0.001, cooldown=6,
                              patience=2, min_lr=1e-6, verbose = 1)
   # tb = TensorBoard(log_dir='./logs', histogram_freq=10, batch_size=64, write_graph=False, write_grads=False, 
    #                 write_images=False, embeddings_freq=0, embeddings_layer_names=None, embeddings_metadata=None)
    return [es, msave, reduce_lr] #, tb]

file_path = ".model_Xception_weights.hdf5"
callbacks = get_callbacks(filepath=file_path, patience=15)


img_generator = ImageDataGenerator(horizontal_flip=True, vertical_flip=True, rotation_range = 10, zoom_range=0.05,
                                  width_shift_range=0.05, height_shift_range=0.05)
```

### Parameters

```python
fit_augmented = True
batch_size = 16
skf = StratifiedKFold(n_splits=5, random_state=5, shuffle = False)
```

```python
def getModel(alpha=1e-3):
    base_model = ResNet50(weights='imagenet', include_top=False)
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dense(512, activation='relu')(x)
    # and a logistic layer -- let's say we have 200 classes
    predictions = Dense(1, activation='sigmoid')(x)

    # this is the model we will train
    model = Model(inputs=base_model.input, outputs=predictions)

    # first: train only the top layers (which were randomly initialized)
    # i.e. freeze all convolutional InceptionV3 layers

   # for layer in model.layers[:105]:
    #    layer.trainable = False
    for layer in model.layers[:]:
        layer.trainable = True

    # compile the model (should be done *after* setting layers to non-trainable)
    #//model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
    mypotim=Adam(lr=alpha, beta_1=0.9, beta_2=0.999, epsilon=1e-06)
    model.compile(loss='binary_crossentropy',
                      optimizer=mypotim,
                      metrics=['accuracy'])
    return model
```

### Creaing and training model

```python
K.clear_session()

#Without denoising, core features.
# train the model on the new data for a few epochs
history_keys = ['val_loss', 'val_acc', 'loss', 'acc']
stat_list = np.array([])
i = 1
for train_index, test_index in skf.split(X_train_all, y_train_all):
    model = getModel()
    print_fold(i)
    i = i + 1 
    # inc_angle_train, inc_angle_test = inc_angle_train_all[train_index], inc_angle_train_all[test_index]
    X_train, X_test = X_train_all[train_index,:,:,:], X_train_all[test_index,:,:,:]
    y_train, y_test = y_train_all[train_index], y_train_all[test_index]
    if (fit_augmented):
        print("Fit augmented data")
        history = model.fit_generator(img_generator.flow(X_train, y_train, batch_size = batch_size),
        steps_per_epoch=len(X_train) / batch_size,
        epochs=150,
        verbose=2,
        validation_data=(X_test, y_test),
        callbacks=callbacks)
        argmin_val_loss = np.argmin(history.history["val_loss"])
        stat_list = np.append(stat_list, [history.history[x][argmin_val_loss] for x in history_keys])
        plot_history(history)
        
    else:
        print("Fit non-augmented data")
        history = model.fit(X_train, y_train,
        batch_size=batch_size,
        epochs=150,
        verbose=2,
        validation_data=(X_test, y_test),
        callbacks=callbacks)
        
        argmin_val_loss = np.argmin(history.history["val_loss"])
        stat_list = np.append(stat_list, [history.history[x][argmin_val_loss] for x in history_keys])
        plot_history(history)
    K.clear_session()
print("*******statistics*******")
stat_list = stat_list.reshape((5,4))
for i in range(4):
    print_stats(history_keys[i], stat_list[range(5),[i]])
```
