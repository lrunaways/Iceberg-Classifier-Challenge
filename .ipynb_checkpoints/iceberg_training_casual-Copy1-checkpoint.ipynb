```python
#фреймворки для построения моделей
import tensorflow as tf
from keras import backend as K
from keras.models import model_from_json
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Input, Flatten, Activation
from keras.layers import GlobalMaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.merge import Concatenate, Add
from keras import initializers
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, Callback, EarlyStopping, ReduceLROnPlateau, TensorBoard
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.inception_v3 import InceptionV3
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K
from keras.utils.vis_utils import model_to_dot
from keras.utils.vis_utils import plot_model

#каллбеки для остановки обучения если модель выходит на плато,
#сохранения модели с минимальным значением функции потерь,
#уменьшения скорости обучения
def get_callbacks(filepath, patience=15):
    es = EarlyStopping('val_loss', patience=patience, mode="min", verbose = 0)
    msave = ModelCheckpoint(filepath, save_best_only=True, verbose = 0)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, verbose=1, cooldown = 5,
                              patience=5, min_lr=5e-6)
    return [es, msave, reduce_lr]
```

```python
#Функции для вывода показателей моделей
def print_fold(n):
    print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")
    print("$$$$$$$$$$$$$$$$$$$", n, " FOLD $$$$$$$$$$$$$$$$$$$$$")
    print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")

def plot_history(history_, split_iter, save_path = None):
    # summarize history for loss
    plt.plot(history_.history['loss'][1:])
    plt.plot(history_.history['val_loss'][1:])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    if save_path is not None:
        plt.savefig("{path}split iteration {i}.png".format(path = dir_path, i = str(split_iter)))
    plt.show()
    
def get_stats_string(stat_name, array):
    s = "\n"\
        "{name}: \n\t{array}\n"\
        "mean {name}: {mean}\n"\
        "str {name}: {std}\n".format(name = stat_name, array = array, mean = array.mean(), std = array.std())
    return s

def output_txt(parameters): 
    stat_list = parameters['stat_list']
    dir_path = parameters['dir_path']
    txt_name = parameters['txt_name']
    history_keys = parameters['history_keys']
    fit_augmented = parameters['fit_augmented']
    with open(dir_path + txt_name, "w") as text_file:
        text_file.write("{model_name}\n\n"\
                        "learning rate: {learning_rate}\n"\
                        "batch_size: {batch_size}"
        .format(**parameters))
        text_file.write("\n\n************Stats************")
        for i in range(len(history_keys)):
            stat_string = get_stats_string(history_keys[i], stat_list[range(len(stat_list)),[i]])
            text_file.write(stat_string)
        text_file.write("*****************************\n")
        if fit_augmented: 
            text_file.write("\nImage generator parameters:\n"\
                    "horizontal_flip = {horizontal_flip}\n"\
                    "vertical_flip = {vertical_flip}\n"\
                    "rotation_range = {rotation_range}\n"\
                    "zoom_range = {zoom_range}\n"\
                    "width_shift_range = {width_shift_range}\n"\
                    "height_shift_range = {height_shift_range}\n".format(**parameters))
            

#def precision_recall_output(y_test, x_test, path = None, model = None):
#    if model is None:
#        if dir_path is None:
#            return -1
#        json_path = path + "model.json"
#        json_file = open(json_path, 'r')
#        loaded_model_json = json_file.read()
#        json_file.close()
#        loaded_model = model_from_json(loaded_model_json)
#        # load weights into new model
#        loaded_model.load_weights(weights_path)
#        model = loaded_model
#    precision, recall, thesholds = precision_recall_curve(y_test,model.predict(x_test, batch_size = 128, verbose = 0))
#    plt.plot(recall,precision)
#    plt.title('precision/recall')
#    plt.xlabel('recall')
#    plt.ylabel('precision')
#    plt.show()
#    return precision, recall, thesholds

def precision_recall_output(y_true, y_pred, path = None):
    precision, recall, thesholds =precision_recall_curve(y_true, y_pred, pos_label=1, sample_weight=None)
    plt.plot(recall,precision)
    plt.title('precision/recall')
    plt.xlabel('recall')
    plt.ylabel('precision')
    plt.axvline(x=0.95, linestyle = "dashed", c = "black")
    plt.axhline(y=0.9, linestyle = "dashed", c = "black")
    if path is not None:
        plt.savefig("{path}precision_recall_curve.png".format(path = path))
        plt.show()
    return precision, recall, thesholds
```

```python
#Создание модели
def getModel(alpha = 1e-3):
    inputs = Input((75, 75, 3), name = "Input") 
    #convolutional layer 1
    x = Conv2D(96, kernel_size=(3, 3),activation=None, name = "conv_1")(inputs)
    x = BatchNormalization(name = "bn_1")(x)
    x = Activation('relu', name = "relu_1")(x)
   # x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_1")(x)
    #convolutional layer 2
    x = Conv2D(96, kernel_size=(3, 3),activation=None, name = "conv_2")(x)
    x = BatchNormalization(name = "bn_2")(x)
    x = Activation('relu', name = "relu_2")(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2),  name = "max_pool_2")(x)
    #convolutional layer 3
    x = Conv2D(128, kernel_size=(3, 3),activation=None, name = "conv_3")(x)
    x = BatchNormalization(name = "bn_3")(x)
    x = Activation('relu', name = "relu_3")(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_3")(x)
      #convolutional layer 4
    x = Conv2D(128, kernel_size=(3, 3),activation=None,  name = "conv_4")(x)
    x = BatchNormalization(name = "bn_4")(x)
    x = Activation('relu', name = "relu_4")(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_4")(x)
   #convolutional layer 5
    x = Conv2D(256, kernel_size=(3, 3),activation=None, name = "conv_5")(x)
    x = BatchNormalization(name = "bn_5")(x)
    x = Activation('relu', name = "relu_5")(x)
    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_5")(x)
  #x = Dropout(0.2, name = "dp_02_5")(x)
    #flatten
    x = Flatten(name = "flatten")(x)
    #Dense 1
    x = Dense(64, activation=None, name = "dense_6")(x)
    x = BatchNormalization(name = "bn_6")(x)
    x = Activation('relu', name = "relu_6")(x)
#    x = Dropout(0.5, name = "dp_04_6")(x)
    #Dense 2
    x = Dense(64, activation=None, name = "dense_7")(x)
    x = BatchNormalization(name = "bn_7")(x)
    x = Activation('relu', name = "relu_7")(x)
    x = Dropout(0.5, name = "dp_04_7")(x)
    # Output 
    output = Dense(1, activation="sigmoid", name = "Output")(x)
    
    model = Model(inputs=inputs, outputs=output)
    optimizer = Adam(lr=alpha, decay=0.0)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

## Training

```python
#Оценка моделей с помощью перекрестной проверки
preds_cv = np.ones_like(y_train_all)*10.0
K.clear_session()

i = 1
for train_index, test_index in skf.split(X_train_all, y_train_all):
    print_fold(i)
    model = getModel(learning_rate)
    X_train, X_test = X_train_all[train_index,:,:,:], X_train_all[test_index,:,:,:] #получение обуч. и контр. данных
    y_train, y_test = y_train_all[train_index], y_train_all[test_index]  
    mean = X_train.mean()      #нормирование данных   
    std = X_train.std()
    X_train = (X_train - mean) / std
    X_test = (X_test - mean) / std
    if (fit_augmented):
        print("Fit augmented data")
        history = model.fit_generator(img_generator.flow(X_train, y_train, batch_size = batch_size),
        steps_per_epoch=len(X_train) / batch_size,
        epochs=150,
        verbose=2,
        validation_data=(X_test, y_test),
        callbacks=callbacks)        
    else:
        print("Fit non-augmented data")
        history = model.fit(X_train, y_train,
        batch_size=batch_size,
        epochs=100,
        verbose=2,
        validation_data=(X_test, y_test),
        callbacks=callbacks)
        
    model.load_weights(weights_path)
    argmin_val_loss = np.argmin(history.history["val_loss"])
    stat_list = np.append(stat_list, [history.history[x][argmin_val_loss] for x in history_keys])
    plot_history(history, split_iter = i, save_path = dir_path)
    preds_cv[test_index] = model.predict(X_test, batch_size=128) 

    K.clear_session()
    i = i + 1

#вывод показателей
print("*******statistics*******")
stat_list = stat_list.reshape((5,4))
    #обьединение показателей на контрольных наборах
for i in range(4):
    print(get_stats_string(history_keys[i], stat_list[range(5),[i]]))    
parameters =   {'model_num': model_num,
                'model_type': model_type,
                'model_name': model_name,
                'dir_path': dir_path,
                'txt_name':txt_name,
                'fit_augmented': fit_augmented,
                'batch_size': batch_size,
                'learning_rate': learning_rate,
                'history_keys': history_keys,
                'stat_list': stat_list,
                'horizontal_flip': horizontal_flip,
                'vertical_flip': vertical_flip,
                'rotation_range': rotation_range,
                'zoom_range': zoom_range,
                'width_shift_range': width_shift_range,
                'height_shift_range': height_shift_range} 
    #сохранение графа модели
model_json = getModel().to_json()
with open(dir_path + "model.json", "w") as json_file:
    json_file.write(model_json)
    #вывод текстового описания графа
stdout = sys.stdout
sys.stdout = open("{}summary.txt".format(dir_path), 'w')
getModel().summary()
sys.stdout = stdout       
    #отрисовка графа  
graph_path = dir_path + 'graph.png'
plot_model(getModel(), to_file= graph_path , show_shapes=True, show_layer_names=True)
    #вывод гиперпараметров и полноты\точности
output_txt(parameters)
precision, recall, thresholds = precision_recall_output(y_train_all, preds_cv, dir_path)
K.clear_session()
```

```python
from sklearn.metrics import precision_recall_curve
# load json and create model
def precision_recall_output(y_test, x_test, dir_path = None, model = None):
    if model is None:
        if dir_path is None:
            return -1
        json_path = dir_path + "model.json"
        json_file = open(json_path, 'r')
        loaded_model_json = json_file.read()
        json_file.close()
        loaded_model = model_from_json(loaded_model_json)
        # load weights into new model
        loaded_model.load_weights(weights_path)
        model = loaded_model
    precision, recall, thesholds = precision_recall_curve(y_test,model.predict(x_test, batch_size = 128, verbose = 0))
    plt.plot(pres,rec)
    plt.show()
    return precision, recall, thesholds
```

```python
def getModel():
    n2 = tf.get_variable("8",[1])
    inputs = Input((75, 75, 3), name = "Input") 
    x = BatchNormalization(name = "bn_0")(inputs)
        #convolutional layer 1
    x = Conv2D(128, kernel_size=(5, 5),activation=None, name = "conv_1")(inputs)
    x = BatchNormalization(name = "bn_1")(x)
    x = Activation('relu', name = "relu_1")(x)
    x = x*n2
#    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_1")(x) ##
        #convolutional layer 2
    x = Conv2D(128, kernel_size=(3, 3),activation=None, name = "conv_2")(x)
    x = BatchNormalization(name = "bn_2")(x)
    x = Activation('relu', name = "relu_2")(x)
#    x = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), name = "max_pool_2")(x)
    #flatten
    x = Flatten(name = "flatten")(x)
    #Dense 1
    x = Dense(96, activation=None, name = "dense_7")(x)
    x = BatchNormalization(name = "bn_7")(x)
    x = Activation('relu', name = "relu_7")(x)
    x = Dropout(0.5, name = "dp_05_7")(x)
    #Dense 2
    x = Dense(96, activation=None, name = "dense_8")(x)
    x = BatchNormalization(name = "bn_8")(x)
    x = Activation('relu', name = "relu_8")(x)
    x = Dropout(0.5, name = "dp_05_8")(x)
    # Output 
    output = Dense(1, activation="sigmoid", name = "Output")(x)
    
    model = Model(inputs=inputs, outputs=output)
    optimizer = Adam(lr=alpha, decay=0.0)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
    
```
